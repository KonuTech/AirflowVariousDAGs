```python
import os
import json
import pandas as pd
import pandasql as ps
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
data = {
    'salesman_id': [11, 12, 13, 11, 12, 12, 11],
    'sale_date': ['2020-04-20', '2020-04-20', '2020-04-22', '2020-04-22', '2020-04-22', '2020-04-21', '2020-04-21'],
    'sale_amount': [12500.00, 12500.00, 11000.00, 11000.00, 22800.00, 9500.00, 31000.00]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

# ROW_NUMBER()

## 1.

```python
pysqldf(
    """
    SELECT
         ROW_NUMBER() OVER( ORDER BY sale_amount) AS row_number
        ,sale_date
        ,salesman_id
        ,sale_amount
    FROM df
    ;
    """
)
```

## 2.

```python
pysqldf(
    """
    SELECT
          ROW_NUMBER() OVER( PARTITION BY sale_date ORDER BY sale_amount ) AS row_number
         ,sale_date
         ,salesman_id
         , sale_amount
    FROM df;
    """
)
```
